## 3. ***Data Wrangling***

### Data Wrangling Code

```python
# checking for missing values in columns
df.isnull().sum()
```

```python
# Convert date columns to datetime format

df['Pickup_Date'] = pd.to_datetime(df['Pickup_Date'])
df['Delivery_Date'] = pd.to_datetime(df['Delivery_Date'])

# Check data types
df.dtypes
```

```python
# Create a new feature: Delivery Time (Days)

df['Delivery_Time_Days'] = (
    df['Delivery_Date'] - df['Pickup_Date']
).dt.days

df.head()
```

```python
df[['Pickup_Date','Delivery_Date','Delivery_Time_Days']].head()
```

####  Check for Outliers

```python
# Numerical Columns

numerical_columns = [
    'Distance_KM',
    'Weight_KG',
    'Cost_USD',
    'Delivery_Time_Days'
]

for column in numerical_columns:
    plt.figure(figsize=(6,3))
    sns.boxplot(x=df[column])
    plt.title(f'Boxplot of {column}')
    plt.show()
```

### What all manipulations have you done and insights you found?

### What all manipulations have you done and insights you found?

During the data wrangling process, the dataset was examined to ensure its quality and readiness for analysis. The following steps were performed:

- Checked for missing values using `isnull().sum()`, and no missing values were found in any column.
- Verified duplicate records using `duplicated().sum()`, and no duplicate entries were present.
- Converted the `Pickup_Date` and `Delivery_Date` columns to datetime format for accurate date-based calculations.
- Created a new feature, `Delivery_Time_Days`, by calculating the difference between the delivery date and pickup date.
- Examined the dataset for outliers using boxplots on numerical variables such as `Distance_KM`, `Weight_KG`, `Cost_USD`, and `Delivery_Time_Days`.

### Insights

- Date conversion enabled the calculation of shipment delivery duration, which is an important performance metric.
- Outliers were observed in shipment weight, distance, and shipping cost. These values were retained because they represent realistic variations in logistics operations rather than data errors.
- The engineered feature `Delivery_Time_Days` provides valuable information for analyzing delivery efficiency, shipment performance, and operational delays.

## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

#### Chart - 1

```python
# Chart 1: Origin vs Shipment Mode

# Group data by Origin and Shipment Mode
data = df.groupby(["Origin", "Shipment_Mode"]).size().unstack(fill_value=0)

# Convert counts to percentages
data_percent = data.div(data.sum(axis=1), axis=0) * 100

# Plot
plt.figure(figsize=(10,6))
ax = data_percent.plot(
    kind="barh",
    stacked=True,
    figsize=(10,6),
    colormap="plasma"
)

plt.title("Percentage Distribution of Shipment Modes by Origin City")
plt.xlabel("Percentage (%)")
plt.ylabel("Origin City")
plt.legend(title="Shipment Mode", bbox_to_anchor=(1.05, 1), loc="upper left")

plt.tight_layout()
plt.show()
```

##### 1. Why did you pick the specific chart?

This bar chart help us to analyze the trends for countries in a better way for each shipment mode.

##### 2. What is/are the insight(s) found from the chart?

- Ground transportation accounts for the largest proportion of shipments in most origin cities.
- Air shipments represent a smaller but consistent share across locations.
- Freight shipments contribute the smallest percentage, indicating they are primarily used for specific logistics requirements.
- The shipment mode distribution appears relatively consistent across origin cities, suggesting standardized transportation practices within the delivery network.

##### 3. Will the gained insights help creating a positive business impact? Are there any insights that lead to negative growth? Justify with specific reason.

Yes. The insights help identify the dominant shipment mode in each origin city, allowing FedEx to allocate transportation resources more efficiently and optimize operational planning.

No major negative trend is observed. However, if a city relies too heavily on a single shipment mode, it could reduce flexibility during disruptions such as bad weather or operational delays.

#### Chart - 11

```python
# Chart 11: Delivery Status Distribution by Shipment Mode

plt.style.use('default')
sns.set_style("whitegrid")

g = sns.FacetGrid(
    df,
    col="Shipment_Mode",
    col_wrap=3,
    height=4,
    aspect=1.2
)

g.map_dataframe(
    sns.countplot,
    x="Delivery_Status",
    order=df["Delivery_Status"].value_counts().index,
    palette="Set2"
)

g.set_titles(col_template="{col_name}")
g.set_axis_labels("Delivery Status", "Number of Shipments")

for ax in g.axes.flatten():
    for p in ax.patches:
        ax.annotate(
            f'{int(p.get_height())}',
            (p.get_x() + p.get_width()/2, p.get_height()),
            ha='center',
            va='bottom',
            fontsize=9
        )
    ax.tick_params(axis='x', rotation=20)

plt.tight_layout()
plt.show()
```

#### Chart - 12

```python
# Chart 12: Monthly Shipment Trend by Shipment Mode

plt.style.use('default')
sns.set_style("whitegrid")

# Create month column
df['Delivery_Month'] = df['Delivery_Date'].dt.month_name()

month_order = ['January','February','March','April','May','June',
               'July','August','September','October','November','December']

shipment_trend = (
    df.groupby(['Delivery_Month', 'Shipment_Mode'])
      .size()
      .unstack(fill_value=0)
      .reindex(month_order)
)

plt.figure(figsize=(14,6))

shipment_trend.plot(
    kind='line',
    marker='o',
    linewidth=2,
    colormap='tab10'
)

plt.title("Monthly Shipment Trend by Shipment Mode", fontsize=15)
plt.xlabel("Month")
plt.ylabel("Number of Shipments")
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title="Shipment Mode", bbox_to_anchor=(1.02,1), loc='upper left')

plt.tight_layout()
plt.show()
```
